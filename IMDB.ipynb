```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r'C:\Users\AYUSHI\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
name=driver.find_elements_by_xpath('//div[@class="lister-list"]/div/div[2]/h3/a')
len(name)
```

```python
Name=[]
for j in name:
    Name.append(j.text)
Name
```

```python
span=driver.find_elements_by_xpath('//div[@class="lister-list"]/div/div[2]/h3/span[2]')
len(span)
```

```python
Year_span=[]
for k in span:
    Year_span.append(k.text)
Year_span
```

```python
gen=driver.find_elements_by_xpath('//div[@class="lister-list"]/div/div[2]/p[1]/span[5]')
len(gen)
```

```python
Genre=[]
for k in gen:
    Genre.append(k.text)
Genre
```

```python
time=driver.find_elements_by_xpath('//div[@class="lister-list"]/div/div[2]/p[1]/span[3]')
len(time)
```

```python
Run_time=[]
for k in time:
    Run_time.append(k.text)
Run_time
```

```python
rating=driver.find_elements_by_xpath('//div[@class="lister-list"]/div/div[2]/div[1]/div[1]/span[2]')
len(rating)
```

```python
Ratings=[]
for k in rating:
    Ratings.append(k.text)
Ratings
```

```python
vote=driver.find_elements_by_xpath('//div[@class="lister-list"]/div/div[2]/p[4]/span[2]')
len(vote)
```

```python
Votes=[]
for k in vote:
    Votes.append(k.text)
Votes
```

```python
imdb=pd.DataFrame()

imdb['Name']= Name
imdb['Year_span ']= Year_span
imdb['Genre']= Genre
imdb['Run_time']= Run_time
imdb['Ratings']= Ratings
imdb['Votes']= Votes
imdb

 
```
