```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r'C:\Users\AYUSHI\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url='https:/www.billboard.com/'
driver.get(url)
```

```python
Url=[]
for i in driver.find_elements_by_xpath("//div[@class='main-menu // u-display-none@desktop-max u-width-100p@desktop-xl-max lrv-u-padding-t-025']/nav/ul/li[1]/a"):
    Url.append(i.get_attribute("href"))
Url
```

```python
driver.get(Url[0])
```

```python
search_btn=driver.find_element_by_xpath('/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[1]/div[1]/div[2]/span/a')
search_btn
```

```python
search_btn.click()
```

```python
name=driver.find_elements_by_xpath('//div[@class="chart-results-list // lrv-u-padding-t-150 lrv-u-padding-t-050@mobile-max"]/div/ul/li[4]/ul/li/h3')
len(name)
```

```python
Song_name=[]
for j in name:
    Song_name.append(j.text)
Song_name
```

```python
Artist=driver.find_elements_by_xpath('//div[@class="chart-results-list // lrv-u-padding-t-150 lrv-u-padding-t-050@mobile-max"]/div/ul/li[4]/ul/li[1]/span')
len(Artist)
```

```python
Artist_name=[]
for k in Artist:
    Artist_name.append(k.text)
Artist_name
```

```python
week=driver.find_elements_by_xpath('//div[@class="chart-results-list // lrv-u-padding-t-150 lrv-u-padding-t-050@mobile-max"]/div/ul/li[4]/ul/li[4]/span')
len(week)
```

```python
Last_week_rank=[]
for k in week:
    Last_week_rank.append(k.text)
Last_week_rank
```

```python
peak=driver.find_elements_by_xpath('//div[@class="chart-results-list // lrv-u-padding-t-150 lrv-u-padding-t-050@mobile-max"]/div/ul/li[4]/ul/li[5]/span')
len(peak)
```

```python
Peak_rank=[]
for k in peak:
    Peak_rank.append(k.text)
Peak_rank
```

```python
weeks=driver.find_elements_by_xpath('//div[@class="chart-results-list // lrv-u-padding-t-150 lrv-u-padding-t-050@mobile-max"]/div/ul/li[4]/ul/li[6]/span')
len(weeks)
```

```python
Weeks_on_board=[]
for k in weeks:
    Weeks_on_board.append(k.text)
Weeks_on_board
```

```python
hot=pd.DataFrame()

hot['Song_name']= Song_name
hot['Artist_name ']= Artist_name
hot['Last_week_rank']= Last_week_rank
hot['Peak_rank']= Peak_rank
hot['Weeks_on_board']= Weeks_on_board
hot

 
```
