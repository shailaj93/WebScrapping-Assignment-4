```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r'C:\Users\AYUSHI\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url='https://archive.ics.uci.edu/'
driver.get(url)
```

```python
Url=[]
for i in driver.find_elements_by_xpath("//span[@class='whitetext']/a"):
    Url.append(i.get_attribute("href"))
Url
```

```python
driver.get(Url[4])
```

```python
name=driver.find_elements_by_xpath('//p[@class="normal"]/b/a')
len(name)
```

```python
Dataset_name=[]
for j in name:
    Dataset_name.append(j.text)
Dataset_name
```

```python
typ=driver.find_elements_by_xpath('//table[2]/tbody/tr/td[2]/p[@class="normal"]')
len(typ)
```

```python
Data_type=[]
for k in typ:
    Data_type.append(k.text)
Data_type
```

```python
task=driver.find_elements_by_xpath('//table[2]/tbody/tr/td[3]/p[@class="normal"]')
len(typ)
```

```python
Task=[]
for k in task:
    Task.append(k.text)
Task
```

```python
tpe=driver.find_elements_by_xpath('//table[2]/tbody/tr/td[4]/p[@class="normal"]')
len(tpe)
```

```python
Attribute_type=[]
for k in tpe:
    Attribute_type.append(k.text)
Attribute_type
```

```python
inst=driver.find_elements_by_xpath('//table[2]/tbody/tr/td[5]/p[@class="normal"]')
len(inst)
```

```python
No_of_instances=[]
for k in inst:
    No_of_instances.append(k.text)
No_of_instances
```

```python
att=driver.find_elements_by_xpath('//table[2]/tbody/tr/td[6]/p[@class="normal"]')
len(att)
```

```python
No_of_attribute=[]
for k in att:
    No_of_attribute.append(k.text)
No_of_attribute
```

```python
year=driver.find_elements_by_xpath('//table[2]/tbody/tr/td[7]/p[@class="normal"]')
len(year)
```

```python
Year=[]
for k in year:
    Year.append(k.text)
Year
```

```python
ML=pd.DataFrame()

ML['Dataset_name']= Dataset_name
ML['Data_type ']= Data_type
ML['Default Task']= Task
ML['Attribute_type']= Attribute_type
ML['No_of_instances']= No_of_instances
ML['No_of_attribute']= No_of_attribute
ML['Year']= Year
ML

 
```
