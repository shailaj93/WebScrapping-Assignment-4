```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r'C:\Users\AYUSHI\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url='https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
driver.get(url)
```

```python
name=driver.find_elements_by_xpath('//table/tbody/tr/td[2][@class="left"]')
len(name)
```

```python
Book_name=[]
for j in name:
    Book_name.append(j.text)
Book_name
```

```python
Author=driver.find_elements_by_xpath('//table/tbody/tr/td[3][@class="left"]')
len(Author)
```

```python
Author_name=[]
for k in Author:
    Author_name.append(k.text)
Author_name
```

```python
sold=driver.find_elements_by_xpathAuthor=driver.find_elements_by_xpath('//table/tbody/tr/td[4][@class="left"]')
len(sold)
```

```python
Volumes_sold=[]
for k in sold:
    Volumes_sold.append(k.text)
Volumes_sold
```

```python
pub=driver.find_elements_by_xpath('//table/tbody/tr/td[5][@class="left"]')
len(pub)
```

```python
Publisher=[]
for k in pub:
    Publisher.append(k.text)
Publisher
```

```python
gen=driver.find_elements_by_xpath('//table/tbody/tr/td[6][@class="last left"]')
len(gen)
```

```python
Genre=[]
for k in gen:
    Genre.append(k.text)
Genre
```

```python
hot=pd.DataFrame()

hot['Book_name']= Book_name
hot['Author_name ']= Author_name
hot['Volumes_sold']= Volumes_sold
hot['Publisher']= Publisher
hot['Genre']= Genre
hot

 
```
